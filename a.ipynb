```python
import pandas as pd    
import matplotlib.pyplot as plt
import seaborn as sns
import streamlit as st

df = pd.read_excel('data/template.xlsx')
fig, ax = plt.subplots(figsize=(8, 4))
sns.barplot(
    data=df,
    x=df['구매총액'],
    y=df.index,
    estimator="mean",   # 평균
    errorbar=None,      # 신뢰구간 제거 (깔끔)
    ax=ax,
)

st.pyplot(fig)
```
